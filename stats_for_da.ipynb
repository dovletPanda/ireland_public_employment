```python
# Choosing the countries, we can start the statistical hypothesis and tests now
# I have selected Italy and Netherland to compare with Ireland;
```

```python
# I will perform one-way ANOVA for those three companies and for above two datasets' Value columns
# Importing library
from scipy.stats import f_oneway
```

```python
# country based splitting
ireland_emp_pub = employ_public.loc[employ_public['Country']=='Ireland']
italy_emp_pub = employ_public.loc[employ_public['Country']=='Italy']
nether_emp_pub = employ_public.loc[employ_public['Country']=='Netherlands']

performance1 = ireland_emp_pub.Value.values
performance2 = italy_emp_pub.Value.values
performance3 = nether_emp_pub.Value.values
 
# Conduct the one-way ANOVA
f_oneway(performance1, performance2, performance3)
```

```python
## Hypothesis Tests...
# I will use 5 different tests
from scipy import stats
```

```python
# 1- One-Sample t-test:
# HYPOTHESIS
## Null Hypothesis (H0): The mean public employment rate is equal to a hypothesized value given as μ = 17 since 2007
## Alternative Hypothesis (Ha): The mean public employment rate is not equal to the hypothesized value (μ ≠ 17) since 2007
###  We choose a confidence level of 99%; that is, we will reject the null hypothesis in favor of the alternative if the p-value is less than 0.01.
stats.ttest_1samp(ireland_emp_pub.Value.values, popmean=17) 
# CONCLUSION As expected, the p-value of 0.019 is not below our threshold of 0.01, so we cannot reject the null hypothesis.
```

```python
# 2- Paired t-test:
# HYPOTHESIS
## Null Hypothesis (H0): There is no significant difference between the means of paired observations (the mean difference = 0),Paired observations: Ireland and Netherlands 2011 average and Ireland and Netherlands 2012 average 
## Alternative Hypothesis (Ha): There is a significant difference between the means of paired observations (e.g., the mean difference ≠ 0).
stats.ttest_rel(performance1, performance3)
# CONCLUSION The p-value of 0.00004 is below our threshold of 0.01, so we can reject the null hypothesis. There is significant difference between the means of paired observations..
```

```python
# 3- Wilcoxon Signed-Rank Test:
# HYPOTHESIS
## Null Hypothesis (H0): There is no significant difference between the medians of paired observations.
## Alternative Hypothesis (Ha): There is a significant difference between the medians of paired observations. 
from scipy.stats import wilcoxon
wilcoxon(performance1, performance3, alternative='greater')
# CONCLUSION The p-value of 0.00003 is below our threshold of 0.01,we can reject the null hypothesis
```

```python
# 4-Two-Sample t-test (Independent t-test):
# HYPOTHESES
## Null Hypothesis (H0): The means of the two groups are equal (μ1 = μ2).
## Alternative Hypothesis (Ha): The means of the two groups are not equal (μ1 ≠ μ2).
stats.ttest_ind(performance1, performance3)
# CONCLUSION As we can predict, the p-value of 0.0000007 is smaller than the threshold and we should reject the null hypothesis..
```

```python
# 5- Chi-Square Test for Independence:
# HYPOTHESES
## Null Hypothesis (H0): There is no association between the country categories and the distribution of public employment rates.
## Alternative Hypothesis (Ha): There is an association between the country categories and the distribution of public employment rates.
stats.chi2_contingency(performance1, correction=False)
# CONCLUSION
```
